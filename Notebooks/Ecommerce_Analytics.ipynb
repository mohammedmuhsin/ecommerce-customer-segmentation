```python
customer_df=(
    df.groupby('customer_unique_id')
    .agg(
        total_orders = ('order_id','nunique'),
        total_revenue = ('revenue','sum'),
        first_purchase_date = ('order_purchase_timestamp','min'),
        last_purchase_date = ('order_purchase_timestamp','max')
    )
    .reset_index()
)
```

```python
import pandas as pd

# choose reference date (max purchase date + 1 day)
reference_date = customer_df['last_purchase_date'].max() + pd.Timedelta(days=1)

# calculate recency
customer_df['recency'] = (reference_date - customer_df['last_purchase_date']).dt.days
```

```python
customer_df['frequency'] = customer_df['total_orders']
```

```python
customer_df['monetary'] = customer_df['total_revenue']
```

```python
customer_df[['recency','frequency','monetary']].describe()
```

```python
customer_df.info()
```

```python
customer_df['R_score']=pd.qcut(
    customer_df['recency'],
    5,
    labels=[5,4,3,2,1,]
)
```

```python
customer_df['F_score']=pd.qcut(
    customer_df['frequency'].rank(method='first'),
    5,
    labels=[1,2,3,4,5]
)
```

```python
customer_df['M_score']=pd.qcut(
    customer_df['monetary'],
    5,
    labels=[1,2,3,4,5]
)
```

```python
customer_df[['R_score','F_score','M_score']].head()
```

```python
customer_df['RFM_score'] = (
    customer_df['R_score'].astype(str)+
    customer_df['F_score'].astype(str)+
    customer_df['M_score'].astype(str)
)
```

```python
customer_df[['RFM_score']].head()
```

```python
def segment_customer(row):
    if row['R_score'] == '5' and row['F_score'] == '5':
        return 'Champions'
    elif row['F_score'] in ['4','5']:
        return 'Loyal Customers'
    elif row['R_score'] == '5':
        return 'New Customers'
    elif row['R_score'] in ['1','2']:
        return 'At Risk'
    else:
        return 'Others'

customer_df['customer_segment'] = customer_df.apply(segment_customer, axis=1)
```

```python
def segment_customer(row):
    if row['R_score'] >= 4 and row['F_score'] >= 3:
        return 'Champions'
    elif row['F_score'] >= 3:
        return 'Loyal Customers'
    elif row['R_score'] >= 4:
        return 'New Customers'
    elif row['R_score'] <= 2:
        return 'At Risk'
    else:
        return 'Potential Loyalists'
customer_df['customer_segment'] = customer_df.apply(segment_customer, axis=1)
```

```python
segment_revenue = (
    customer_df
    .groupby('customer_segment')['monetary']
    .sum()
    .sort_values(ascending=False)
)

segment_revenue
```

```python
segment_avg_value = (
    customer_df
    .groupby('customer_segment')['monetary']
    .mean()
    .sort_values(ascending=False)
)

segment_avg_value
```

```python
segment_recency = (
    customer_df
    .groupby('customer_segment')['recency']
    .mean()
    .sort_values()
)

segment_recency
```

```python
segment_frequency = (
    customer_df
    .groupby('customer_segment')['frequency']
    .mean()
    .sort_values(ascending=False)
)

segment_frequency
```

```python
rfm_summary = (
    customer_df
    .groupby('customer_segment')
    .agg(
        customers=('customer_unique_id', 'count'),
        avg_recency=('recency', 'mean'),
        avg_frequency=('frequency', 'mean'),
        avg_monetary=('monetary', 'mean'),
        total_revenue=('monetary', 'sum')
    )
    .sort_values('total_revenue', ascending=False)
)

rfm_summary
```

```python
monthly_sales = monthly_sales.reset_index()
monthly_sales['order_month'] = monthly_sales['order_month'].dt.to_timestamp()
```

```python
monthly_sales.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.plot(monthly_sales['order_month'],
         monthly_sales['payment_value'],
         color='blue',
         linewidth = 2
        )
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
monthly_sales['moving_avg_3']=monthly_sales['payment_value'].rolling(window=3).mean()
```

```python
monthly_sales.head()
```

```python
plt.figure()

plt.plot(
    monthly_sales['order_month'],
    monthly_sales['payment_value'],
    label='Actual Sales'
)

plt.plot(
    monthly_sales['order_month'],
    monthly_sales['moving_avg_3'],
    label='3 Month Moving Average'
)

plt.title('Monthly Sales with Moving Average')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```
